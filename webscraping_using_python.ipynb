```python
import requests
from bs4 import BeautifulSoup
```

```python
booklist={"Book Name":[],"writer":[],"Book Text":[],"Genre":[]}
for i in range(1,10+1):
  url="https://banglakobita.com.bd/গল্প-উপন্যাস/page/"+str(i)+"/"
  response=requests.get(url)
  soup=BeautifulSoup(response.text,'html.parser')
  book_name=soup.find_all('h2',class_="sek-pg-title")
  book=soup.find_all('div',class_="sek-pg-category")
  book_text=soup.find_all('figure',class_="sek-pg-thumbnail")
  # print(book_text)


  for j in range(len(book_name)):
    book_name_writer=book_name[j].text.strip().split('– ')
    # print(i,book_name_writer)
    if len(book_name_writer)>=2:
      booklist['Book Name'].append(book_name_writer[0])
      booklist['writer'].append(book_name_writer[1])
    else:
      book_name_writer=book_name_writer[0].split(' ')
      booklist['Book Name'].append(book_name_writer[-1])
      booklist['writer'].append(book_name_writer[0])
    genre=book[j].text.split('/')
    booklist['Genre'].append(genre)

    ##book text section by getting url
    book_text_url=str(book_text[j].find('a')['href'])
    res=requests.get(book_text_url)
    soup2=BeautifulSoup(res.text,'html.parser')
    x=soup2.find('div',class_='entry-content clearfix')
    y=x.find_all('p')
    full_text=''
    for k in range(2,len(y)):

      full_text+=(y[k].text+' ')

    booklist['Book Text'].append(full_text)

print(booklist)

```

```python
import pandas as pd
df=pd.DataFrame(booklist)
df
```
